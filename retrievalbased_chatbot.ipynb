```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
data = pd.read_csv(r'C:\RAIN\chatbot\Conversations.csv')
```

```python
data.head()
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(data['Sentences'])#learns the sentences goes through your sentences to understand
word_vectors = vectorizer.transform(data['Sentences'])

```

```python
word_vectors
```

```python
pd.DataFrame(word_vectors.toarray(),
             columns = vectorizer.get_feature_names_out())
```

```python
def get_response(question):
    response = None
    question_vector =vectorizer.transform([question])
    #print(question_vector)
    similarity_scores = cosine_similarity(question_vector,word_vectors)[0]
    #print(similarity_scores)
    max_score = np.max(similarity_scores)
    #print(max_score)
    threshold = 0.6
    if max_score >= threshold:
        idx = np.argmax(similarity_scores)# argmax returns an index
        #print(idx)
        #print(data.iloc[idx]['Responses'])
        response = data.iloc[idx]
    else:
        with open('logs','a') as f:
            f.write(question+ '\n')
    return response
```

```python
get_response('Good Morning')#similarity of our new data with the existing datasset
```

```python
get_response('what date is it?')
```

```python
while True:
    quest = input('Me:')
    print(f"Me: {quest}")
    result = get_response(quest)
    try:
        intent =result['Intent']
        reply = result['Responses']
        print(f"Bot: {reply}")
        if (intent.strip() == 'goodbye') or (intent.strip() == 'good bye'):
            break
    except Exception as err:
        print(err)
        print(f'Bot: Please, can you rephrase your question?')
```
